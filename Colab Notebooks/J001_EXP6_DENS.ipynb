```python
import random as rand
```

```python
def euclid(m,n):
  if n==0:
    return m
  else:
    r= m%n
    return euclid(n,r)
```

```python
#Program to find Multiplicative inverse
def exteuclid(a,b):
  r1=a
  r2=b
  s1= int(1)
  s2= int(0)
  t1= int(0)
  t2= int(1)
  while r2>0:
    q= r1//r2
    r= r1-q*r2
    r1= r2
    r2 = r
    s= s1-q*s2
    s1=s2
    s2=s
    t= t1-q*t2
    t1=t2
    t2=t
  if t1<0:
    t1= t1%a
  return (r1,t1)
```

```python
l = []            #initialize an empty list
for num in range(1, 251):         #range can be 1 to x, when x is any +ve int
   # all prime numbers are greater than 1
   if num > 1:
       for i in range(2, num):              #if divisible by i then not a prime no
           if (num % i) == 0:
               break
       else:
           l.append(num)       #append the number not divisible 
```

```python
data = rand.sample(l,2)           #pick two random prime number from the list of prime numbers
p = data[0] 
q = data[1]
```

```python
print("The first prime number: ", p )
print("The second prime number: ", q )
```

```python
PT= input("Enter the plain text to be encrypted: ")
```

```python
plain_text = []
for i in range(len(PT)):
  plain_text.append(ord(PT[i]))
```

```python
n= p*q
print(n)
Pn= (p-1)*(q-1)
```

```python
# Generate encryption key in range 1<e<Pn
key= []
for i in range(2,Pn):
  gcd= euclid(Pn,i)
  if gcd==1:
    key.append(i)
print("All possible values of encryption key are: ", key)
```

```python
# Select an encryption key from the above list
e= int(13)
```

```python
#Obtain inverse of encryption key in Z_Pn
r, d= exteuclid(Pn,e)
if r==1:
  d=int(d)
  print("decryption key is: ", d)
else:
  print("Multiplicative inverse for the given encryption key does not exist. Choose a different encrytion key ")
```

```python
ct = []
for i in range(len(plain_text)):
  a=(pow(plain_text[i],e))%n
  ct.append(a)
```

```python
dt = []
for i in range(len(plain_text)):
  b=(pow(ct[i],d))%n
  dt.append(b)
```

```python
decrypted_text = []
for i in range(len(dt)):
  decrypted_text.append(chr(dt[i]))
```

```python
print("For plain text = ", PT, ", Cipher text = ", ct, "Decrypted text = ", decrypted_text)
```
