**AIM: To generate subkeys for 16 rounds of DES**

**Importing libraries**

```python
import numpy as np
import random
```

**Generating Key of size 64 elements. Here, insead of bits, integers are taken to understand the permutaion and shifting operations taking place.**

```python
key= np.arange(1,65,1)
print(key)
```

**Initializing permutaion choice 1 matrix**

```python
Perm1= [57,49,41,33, 25,17,9,1,58,50,42,34,26,18,10,2,59,51,43,35,27,19,11,3,60,52,44,36,63,55,47,39,31,23,15,7,62,54,46,38,30,22,14,6,61,53,45,37,29,21,13,25,28,20,12,4]
```

**Performing permutation and parity drop operations**

```python
key1=[]
for i in range(0,56):
  x= Perm1[i]
  x=int(x)
  temp= key[x]
  key1.append(temp)
print(key1)
```

```python
leftKey1= key1[0:28]
rightKey1= key1[28::]
```

**Performing left rotation of each round key according to its round number**

```python
keyrot= np.zeros([16,56],dtype=int)
for i in range(0,16):
  temp=[]
  if i== 0 or i==1 or i==8 or i==15:
    leftKey1= np.roll(leftKey1,-1)
    rightKey1= np.roll(rightKey1,-1)
  else:
    leftKey1= np.roll(leftKey1,-2)
    rightKey1= np.roll(rightKey1,-2)
  temp= np.concatenate((leftKey1,rightKey1))
  keyrot[i,:]= temp
```

**Initializing permutaion choice 2 matrix**

```python
#Initialize permutaion choice 2 matrix
Perm2= [14,17,11,24,1,5,3,28,15,6,21,10,23,19,12,4,26,8,16,7, 27,20,13,2,41,52,31,37,47,55,30,40,51,45,33,48,44,49,39,56,34,53,46,42,50,36,29,32]
```

**Performing permutation and compression**

```python
#Performs permutation and compression
keyround=np.zeros([16,48], dtype=int)
for i in range(0,16):
  tempkey= keyrot[i,:]
  key2=[]
  for k in range(0,48):
    y=Perm2[k]-1
    temp=tempkey[y]
    key2.append(temp)
  keyround[i,:]=key2
  print("Key for round",i,"= ", tempkey)
```
